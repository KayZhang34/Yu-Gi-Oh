```python
import json
import requests
import pandas as pd

url_cards = "https://db.ygoprodeck.com/api/v7/cardinfo.php"
url_sets = "https://db.ygoprodeck.com/api/v7/cardsets.php"

header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}


cards = requests.get(url_cards)
sets = requests.get(url_sets)

df_cards = pd.DataFrame.from_dict(cards.json()['data'])
df_sets = pd.DataFrame.from_dict(sets.json())

#adding missing sets/consolidating set names
df_cards.at[df_cards['card_sets'][df_sets['card_name'] == '2020 Tin of Lost Memories'], 'set_name'] ='2020 Tin of Lost Memories Mega Pack'

df_add = {'set_name': 'Yu-Gi-Oh! TCG Remote Duel Extravaganza participation card', 'set_code': '21CC', 'num_of_cards': 1, 'tcg_date': '2021-05-31'}
df_sets = df_sets.append(df_add, ignore_index = True)


#removing cards in the database that were not released in the TCG
df_tcg_cards=df_cards[df_cards['card_sets'].notna()]
```

```python
for row in df_tcg_cards['card_sets']:
    print(row)
    print(df_sets.loc[df_sets['set_name']==row[0]['set_name'],'tcg_date'])
    date = df_sets.loc[df_sets['set_name']==row[0]['set_name'],'tcg_date'].item()
    print(date)
    dates.append(date)
```
